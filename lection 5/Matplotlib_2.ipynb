## График точек

Зададим наборы данных с помощью массивов, состоящих из случайных целых чисел

```python
x_scatter_1 = np.random.randint(0, 50, 100)
x_scatter_2 = np.random.randint(50, 100, 100)
y_scatter_1 = np.random.randint(0, 50, 100)
y_scatter_2 = np.random.randint(50, 100, 100)
z_scatter_1 = np.random.randint(0, 50, 100)
z_scatter_2 = np.random.randint(50, 100, 100)
```

Построим трёхмерный график точек

```python
fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_scatter_1, y_scatter_1, z_scatter_1,
           marker='x', c='salmon', label='class 1')
ax.scatter(x_scatter_2, y_scatter_2, z_scatter_2,
           marker='o', c='deepskyblue', label='class 2')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.zaxis.labelpad = 0

plt.legend()

plt.tight_layout()

plt.show()
```

## Построение трёхмерных графиков кривых, заданных параметрически

Зададим кривую

```python
t = np.linspace(0, 8*np.pi, 1000)

x = t*np.cos(t)
y = t*np.sin(t)
z = t
```

Отобразим график

```python
fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, c='deepskyblue', ls='--', lw=3)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.zaxis.labelpad = -1

plt.title('Параметрическая кривая')

plt.tight_layout()

plt.show()
```

## Построение поверхностей

Создание сетки

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)
print(x.shape, '\n')
print(x, '\n')
print(y.shape, '\n')
print(y)
```

Вычисление значений функции в узлах сетки

```python
z = x**2-y**2
```

Отобразим поверхность

```python
fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap="plasma")

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.zaxis.labelpad = -1

plt.title(r'$z=x^2-y^2$')

plt.tight_layout()

plt.show()
```

## Построение параметрически заданных поверхностей

Зададим поверхность как 
$$
\begin{cases}
x(u,v)=\left(1+\frac{v}{2}\cos\frac{v}{2}\right)\cos u,\\
y(u,v)=\left(1+\frac{v}{2}\cos\frac{v}{2}\right)\sin u,\\
z(u,v)=\frac{v}{2}\sin\frac{v}{2},
\end{cases}$$
где $0\leq u<2\pi$ и $-1\leq v \leq 1$

```python
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(-1, 1, 100)

u, v = np.meshgrid(u, v)

x = (1 + v/2*np.cos(u/2)) * np.cos(u)
y = (1 + v/2*np.cos(u/2)) * np.sin(u)
z = v/2 * np.sin(u/2)
```

Отобразим поверхность

```python
fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, alpha=0.75, cmap='Spectral')

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

plt.tight_layout()

plt.axis('off')

plt.show()
```
