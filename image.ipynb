```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import PIL.ImageOps

image = Image.open('ulm.jpg')
image = image.resize((512, 512))
```

```python
matr=np.array(image.convert('L'))
```

```python
grad_z=np.gradient(matr)
dzdi=np.abs(grad_z[0])
dzdj=np.abs(grad_z[1])
matr_dz=(dzdi+dzdj)/2

matr_dz=np.where(matr_dz < 10, 0, 255)
```

```python
image_dzdi=Image.fromarray(dzdi).convert('L')
image_dzdj=Image.fromarray(dzdj).convert('L')
image_matr_dz=Image.fromarray(matr_dz).convert('L')

```

```python
new_image = Image.new('RGB', (2*image.size[0],image.size[1]), (250,250,250))
# image_inv=PIL.ImageOps.invert(image_dzdi)
```

```python
new_image.paste(image,(0,0))
new_image.paste(image_matr_dz,(image_matr_dz.size[0],0))

```
