```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
```

```python
df=pd.read_csv('data\penguins_size.csv')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
adelie_culmen_sizes=df[['culmen_length_mm','culmen_depth_mm']][df['species']=='Adelie'].to_numpy()
chinstrap_culmen_sizes=df[['culmen_length_mm','culmen_depth_mm']][df['species']=='Chinstrap'].to_numpy()
gentoo_culmen_sizes=df[['culmen_length_mm','culmen_depth_mm']][df['species']=='Gentoo'].to_numpy()
```

```python
[adelie_x,adelie_y]=adelie_culmen_sizes.T
[chinstrap_x,chinstrap_y]=chinstrap_culmen_sizes.T
[gentoo_x,gentoo_y]=gentoo_culmen_sizes.T
```

```python
plt.scatter(adelie_x,adelie_y,marker='x',c='red')
plt.scatter(chinstrap_x,chinstrap_y,marker='o',c='blue')
plt.scatter(gentoo_x,gentoo_y,marker='v',c='green')
```

```python
adelie_body_mass=df['body_mass_g'][df['species']=='Adelie'].to_numpy()
chinstrap_body_mass=df['body_mass_g'][df['species']=='Chinstrap'].to_numpy()
gentoo_body_mass=df['body_mass_g'][df['species']=='Gentoo'].to_numpy()
```

```python
adelie_body_mass
```

```python
plt.scatter(adelie_x,adelie_y,marker='x',c=adelie_body_mass)
plt.scatter(chinstrap_x,chinstrap_y,marker='o',c=chinstrap_body_mass)
plt.scatter(gentoo_x,gentoo_y,marker='v',c=gentoo_body_mass)
plt.colorbar()
```

```python
from matplotlib.ticker import ScalarFormatter
class ScalarFormatterClass(ScalarFormatter):
        def _set_format(self):
            self.format = "%1.1f"
```

```python
matplotlib.rcParams['font.size']=16
yScalarFormatter = ScalarFormatterClass(useMathText=True)
yScalarFormatter.set_powerlimits((0,0))

plt.figure(figsize=(18,6))

plt.subplot(131)
plt.scatter(adelie_x,adelie_y,marker='x',c=adelie_body_mass,cmap='inferno')

plt.colorbar(format=yScalarFormatter)


plt.subplot(132)

plt.scatter(chinstrap_x,chinstrap_y,marker='o',c=chinstrap_body_mass,cmap='nipy_spectral')
plt.colorbar(format=yScalarFormatter)

plt.subplot(133)

plt.scatter(gentoo_x,gentoo_y,marker='v',c=gentoo_body_mass)
plt.colorbar(format=yScalarFormatter)
```
