<h1 style="text-align: center;">Seaborn</h1>

Импортируем необходимые библиотеки

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Откроем набор данных с помощью **pandas**

```python
df = pd.read_csv('data\penguins_size.csv')
df.dropna(inplace=True)
df.drop(df.index[df['gender'] == '.'], inplace=True)
```

```python
df.head()
```

Для красоты попросим seaborn установить тему по умолчанию с помощью *sns.set_theme()* и отобразим типовые графики

```python
sns.set_theme()
```

## ScatterPlot

```python
sns.scatterplot(df, x='flipper_length_mm', y='body_mass_g')
plt.show()
```

## LinePlot


По умолчанию график показывает оценку центрального тенда и доверительный интервал для этой оценки.

```python
sns.lineplot(data=df, x="flipper_length_mm",  y="body_mass_g")
plt.show()
```

## StripPlot

```python
sns.stripplot(data=df, x="body_mass_g", y="species")
plt.show()
```

## BarPlot

```python
sns.barplot(data=df, x="species", y="body_mass_g")
plt.show()
```

## CountPlot

```python
sns.countplot(x='island', data=df)
plt.show()
```

## HistPlot

```python
sns.histplot(df["body_mass_g"], binwidth=200, kde=True, stat='density')
plt.show()
```

## ECDfPlot

```python
sns.ecdfplot(data=df, x="body_mass_g")
plt.show()
```

## BoxPlot

```python
sns.boxplot(x='species', y='flipper_length_mm', data=df)
plt.show()
```

## Pairplot

```python
sns.pairplot(df, hue="species")
plt.show()
```

## ViolinPlot

```python
sns.violinplot(x=df['island'], y=df['culmen_length_mm'],
               hue=df['gender'], split=True)
plt.show()
```
